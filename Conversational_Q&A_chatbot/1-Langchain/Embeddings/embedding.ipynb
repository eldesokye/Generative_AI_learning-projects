```python
### example  first loader the data from a txt file 

from langchain_community.document_loaders import TextLoader

loader = TextLoader("speech.txt")
text = loader.load()
text
```

```python
## then we will split the file to chunks using RecursiveCharactertextsplitter

from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(chunk_size = 500 , chunk_overlap = 20 )
final_docs = splitter.split_documents(text)
final_docs
```

```python
# third we will create a embedd model 

from langchain_community.embeddings import OpenAIEmbeddings


embeddings_1024 = OpenAIEmbeddings(
    model = "openai/text-embedding-3-small",
    
) 

```

```python
## vector Embedding and vector storeDB 
from langchain_community.vectorstores import Chroma

df = Chroma.from_documents(final_docs , embeddings_1024)
df
```

```python
# lets test it for search in the query 

query = "It will be all the easier for us to conduct ourselves as belligerents in a high spirit of right"

retrived_results = df.similarity_search(query)
print(retrived_results)
```
