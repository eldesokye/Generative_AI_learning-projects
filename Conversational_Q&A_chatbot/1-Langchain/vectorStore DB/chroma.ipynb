## Chroma
* chroma is a Ai-native open-source vector database focused on developer productivity and hapiness . 
* chroma is licensed under Apache 2.0 

```python
## building a sample vectordb 
from langchain_chroma import Chroma 
from langchain_community.document_loaders import TextLoader
from langchain_community.embeddings import OllamaEmbeddings 
from langchain_text_splitters import CharacterTextSplitter
```

```python
## load data 
loader = TextLoader('speech.txt')
data = loader.load()
data
```

```python
## split 
spletting =CharacterTextSplitter(chunk_size=1000 , chunk_overlap = 50)
split_data = spletting.split_documents(data)

split_data
```

```python
# embedding and vector store 

embeding = OllamaEmbeddings( model = "mxbai-embed-large")
vectordb= Chroma.from_documents(documents=split_data, embedding = embeding)

vectordb
```

```python
## qurey it 

query = "what does the speaker believe is the main reason the United States should enter the war?"

docs = vectordb.similarity_search(query)

docs[0].page_content
```

```python
## saving to the disk 

vectordb= Chroma.from_documents(documents=split_data, embedding = embeding,persist_directory="./chroma_db")



```

```python
## load from the disk 

db2 = Chroma(persist_directory="./chroma_db",embedding_function=embeding)
docs = db2.similarity_search(query)
docs[0].page_content
```

```python
## similarity Search with Scores 

docs = vectordb.similarity_search_with_score(query)
docs
```

```python
### Retriever option 

retriever = vectordb.as_retriever()

retriever.invoke(query)[0].page_content
```
